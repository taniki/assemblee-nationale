```python
import json
from glob import glob
import pandas as pd
```

```python
def get_acteur(fn):
    data = json.load(open(fn))
    return {
        'uid': data['acteur']['uid']['#text'],
        'nom': data['acteur']['etatCivil']['ident']['nom'],
        'prenom': data['acteur']['etatCivil']['ident']['prenom']
    }

acteurs = pd.DataFrame.from_records([ get_acteur(fn) for fn in glob('refs/PA*.json') ])

acteurs
```

```python
acteurs.to_csv('out/acteurs.csv', index=False)
```

```python
def get_organe(fn):
    data = json.load(open(fn))['organe']
    return data

organes = pd.DataFrame.from_records([ get_organe(fn) for fn in glob('refs/PO*.json') ])

organes
```

```python
organes.to_csv('out/organes.csv', index=False)
```
