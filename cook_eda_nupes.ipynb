```python
import pandas as pd
```

```python
organes = (pd.read_csv('out/organes.csv'))
votes = pd.read_csv('out/votes.csv')
```

```python
nupes = (
    organes
    .query('libelle.str.contains("NUPES") or libelle.str.contains("France insoumise")')
)

nupes
```

```python
alignements = (
    votes
    .query('organe.isin(@nupes.uid)')
    .groupby(['scrutin', 'organe'])
    .agg({
        'groupe_majorite': 'first'
    })
    .reset_index()
    .assign(
        organe = lambda df: df.join(nupes.set_index('uid'), on='organe')[['libelle']]
    )
    .pivot_table(
        index="scrutin",
        columns="organe",
        values="groupe_majorite",
        aggfunc= lambda x: x
    )
    # .assign(
    #     alignement= lambda df: df.eq(df.iloc[:, 0], axis=0).all(1)
    # )
)

alignements
```

```python
alignements
```

```python
votes_alignes = (
    alignements
    .reset_index()
    .melt(id_vars=['scrutin'])
    .set_index(['scrutin', 'organe'])
    #[['value']]
    # je suis vraiment pas fier de moi là
    .assign(
        a= lambda df: df.value,
        b= lambda df: df.value,
        c= lambda df: df.value
    )
    .set_axis(alignements.columns, axis=1)
    .eq(alignements, axis=0)
    .reset_index()
    .drop(columns=['scrutin'])
    .groupby('organe')
    .sum()
)

votes_alignes
```

```python
import numpy as np
```

```python
(
    votes_alignes
    .div(
        votes_alignes.max(), axis=0
    )
    .replace(1, np.nan)
    .style
        .format('{:,.1%}'.format)
        .background_gradient(
            cmap='PiYG',
            vmin=0.5,
            vmax=1
        )
)
```

```python
(
    votes_alignes
    .div(
        votes_alignes.max(), axis=0
    )
    .replace(1, np.nan)
    .mean(axis=1)
    .to_frame()
)
```
