```python
import pandas as pd
```

```python
reunions = pd.read_csv('out/reunions.csv')
organes = pd.read_csv('out/organes.csv')
acteurs = pd.read_csv('out/acteurs.csv')
```

```python
acteurs
```

```python
reunions_full = (
    reunions
    .join(organes.set_index('uid'), on='organeReuniRef')
    .join(acteurs.set_index('uid'), on='demandeur', rsuffix='demandeur_')
)

reunions_full
```

```python
(
    reunions_full
    .query('~organeReuniRef.isna()')
    .libelle
    .value_counts()
)
```
