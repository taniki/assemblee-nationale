```python
import pandas as pd
```

```python
votes = pd.read_csv('out/acteurs_vecteur-scrutins_pca.csv', index_col='acteurRef')

votes
```

```python
renaissance = (
    votes
    .query('libelle == "Renaissance"')
    [[ c for c in votes.columns if c.startswith('VT') ]]
    .mean()
    # .to_frame()
    # .T
    .round()
)

renaissance
```

```python
(
    votes
    .query('libelle == "Renaissance"')
    [[ c for c in votes.columns if c.startswith('VT') ]]
    .eq(renaissance)
    .mean(axis=1)
    .mean()
)
```

```python
(
    votes
    .query('libelle == "La France insoumise - Nouvelle Union Populaire écologique et sociale"')
    [[ c for c in votes.columns if c.startswith('VT') ]]
    .eq(renaissance)
    .mean(axis=1)
    .mean()
)
```
