```python
import pandas as pd
import numpy as np
import json
import pandas as pd
from glob import glob
import tqdm
```

```python
import dagstermill
```

```python
def get_reunions(filename):
    return json.load(open(filename))['reunion']

reunions_json = [ get_reunions(f) for f in glob('data/reunions/json/**/RU*.json') ]

len(reunions_json)
```

```python
reunions_ = (
    pd.DataFrame.from_dict(pd.json_normalize(reunions_json))
)

reunions_.columns
```

```python
reunions_.query('~`participants.participantsInternes.participantInterne`.isna()')#[['participants.participantsInternes.participantInterne']]
```

```python
reunions.to_csv('out/reunions.csv', index=False)
```
